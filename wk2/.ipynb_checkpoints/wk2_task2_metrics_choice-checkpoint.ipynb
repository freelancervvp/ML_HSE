```python
import pandas
import numpy as np
from sklearn.datasets import load_boston
df = load_boston()
print(df.data)
print(df.target)
print(df.data.shape)
```

```python
from sklearn.preprocessing import scale
df_feat_std = scale(df.data)
print(df_feat_std)
df_y = df.target
print(df_y)
```

```python
from sklearn import cross_validation

kf = cross_validation.KFold(len(df_y), n_folds=5, shuffle=True, random_state=42)
print(len(kf))
print(kf) 
```

```python
from sklearn import neighbors

scores = numpy.zeros((200, 2))

i=0
for p1 in np.linspace(1, 10, num=200):
        clf = neighbors.KNeighborsRegressor(n_neighbors=5, weights='distance', metric='minkowski', p = p1)
        cv_scores = cross_validation.cross_val_score(clf, df_feat_std, df_y, cv=kf, scoring='mean_squared_error')
        scores[i,:] = [p1, cv_scores.mean()]
        i=i+1
        
print(cv_scores)
#print(cv_scores.shape)
print(scores)
#print(scores.argmax())
print(scores[:,1].max())
#print(scores.argmin())
print(scores[:,1].min())
```
