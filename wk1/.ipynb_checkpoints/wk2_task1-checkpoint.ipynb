```python
import pandas
import numpy
df = pandas.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", names = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline'], index_col=False)
df.head(10)
```

```python
df_feat = df.iloc[:,1:]
df_feat.head(10)
```

```python
df_y = df.iloc[:,0]
df_y.head(10)
```

```python
len(df_y.index)
```

```python
from sklearn import cross_validation
kf = cross_validation.KFold(len(df_y.index), n_folds=5, shuffle=True, random_state=42)
len(kf)
print(kf)  

for train_index, test_index in kf:
    print("TRAIN:", train_index, "TEST:", test_index)
    #X_train, X_test = df_feat.iloc[train_index,:], df_feat.iloc[test_index,:]
    #y_train, y_test = df_y.iloc[train_index], df_y.iloc[test_index]
    #print(y_test)
```

```python
from sklearn import neighbors
from sklearn.metrics import accuracy_score

scores = numpy.zeros(50)

for n_neighbors in range(1, 51):
    #for train_index, test_index in kf:
        #print(n_neighbors)
        clf = neighbors.KNeighborsClassifier(n_neighbors, weights='uniform')
        #clf.fit(df_feat.iloc[train_index,:], df_y.iloc[train_index])
        #print(df_feat.iloc[test_index,:])
        #z = clf.predict(df_feat.iloc[test_index,:])
        cv_scores = cross_validation.cross_val_score(clf, df_feat, df_y, cv=kf, scoring='accuracy')
        #accuracy_score(df_y[test_index], z)
        scores[n_neighbors-1] = cv_scores.mean()
    
print(scores)
print(scores.argmax())
print(scores.max())
```

```python
from sklearn.preprocessing import scale
df_feat_std = scale(df_feat)

#print(df_feat_std[:10,:])

scores = numpy.zeros(50)

for n_neighbors in range(1, 51):
    #for train_index, test_index in kf:
        #print(n_neighbors)
        clf = neighbors.KNeighborsClassifier(n_neighbors, weights='uniform')
        #clf.fit(df_feat.iloc[train_index,:], df_y.iloc[train_index])
        #print(df_feat.iloc[test_index,:])
        #z = clf.predict(df_feat.iloc[test_index,:])
        cv_scores = cross_validation.cross_val_score(clf, df_feat_std, df_y, cv=kf, scoring='accuracy')
        #accuracy_score(df_y[test_index], z)
        scores[n_neighbors-1] = cv_scores.mean()
    
print(scores)
print(scores.argmax())
print(scores.max())
```
