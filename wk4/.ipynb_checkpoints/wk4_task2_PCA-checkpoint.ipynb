```python
import pandas as pd
import numpy as np
close_prices = pd.read_csv("close_prices.csv")#, header = None, names = ['y', 'x1', 'x2'])
print(close_prices.shape)
close_prices.head(10)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10)
pca.fit(close_prices.iloc[:, 1:])
```

```python
print(pca.explained_variance_ratio_)
```

```python
print(pca.components_.shape)
print(pca.components_[0,:])

print(np.max(pca.components_[0,:]), np.argmax(pca.components_[0,:]), close_prices.columns[np.argmax(pca.components_[0,:])+1])
```

```python
close_prices_pca = pca.transform(close_prices.iloc[:, 1:])
print(close_prices_pca.shape)
print(close_prices_pca[:10,:1])
```

```python
djia_index = pd.read_csv("djia_index.csv")#, header = None, names = ['y', 'x1', 'x2'])
print(djia_index.shape)
djia_index.head(10)
```

```python
print(close_prices_pca[:,0].shape)
print(djia_index['^DJI'].shape)
pearson = np.corrcoef(close_prices_pca[:,0], djia_index['^DJI'])
print(pearson)
```
