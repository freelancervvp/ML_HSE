```python
import pandas as pd
import numpy as np
abalone = pd.read_csv("abalone.csv")#, header = None, names = ['y', 'x1', 'x2'])
print(abalone.shape)
abalone.head(10)
```

```python
#abalone['Sex'] = abalone['Sex'].map(lambda x: 1 if x == 'M' else (-1 if x == 'F' else 0))
abalone['Sex'] = [1 if x == 'M' else (-1 if x == 'F' else 0) for x in abalone['Sex']]
abalone.head(10)
```

```python
abalone_features = abalone.iloc[:,:-1]
abalone_features.head(10)
```

```python
abalone_target = abalone.iloc[:,-1:]
abalone_target.head(10)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import KFold
from sklearn.metrics import r2_score
from sklearn.cross_validation import cross_val_score

kf = KFold(len(abalone_target), n_folds=5, shuffle=True, random_state=1)
scores = np.zeros(50)

for n_trees in range(1, 51):
    clf = RandomForestRegressor(n_estimators = n_trees, random_state = 1)
    cv_scores = cross_val_score(clf, abalone_features, abalone_target, cv = kf, scoring='r2_score')
    scores[n_trees-1] = cv_scores.mean()
    print(cv_scores)
```
